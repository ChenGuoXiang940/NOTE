### 20. è©¦ç”±éµç›¤è¼¸å…¥ä¸€å€‹æ•´æ•¸ï¼Œç„¶å¾Œåˆ¤æ–·è¼¸å…¥çš„æ•¸æ˜¯å¹¾å€‹ä½æ•¸çš„æ•´æ•¸ã€‚ä¾‹å¦‚è‹¥è¼¸å…¥23983ï¼Œå‰‡è¼¸å‡º â€œ5å€‹ä½æ•¸çš„æ•´æ•¸â€ã€‚ï¼ˆæç¤ºï¼šä¸€å€‹æ•´æ•¸ç”¨æ•´æ•¸é™¤æ³•é™¤ä»¥10ï¼Œå…¶ä½æ•¸æœƒå°‘ä¸€ä½ï¼‰ã€‚

### 21. æœ‰æ™‚è®€é€²ä¾†çš„æ•¸å­—æœƒåŒ…å«æœ‰åƒåˆ†ä½ç¬¦è™Ÿï¼ˆå³é€—è™Ÿï¼Œå¦‚13,988æˆ–1,661,231ï¼‰ã€‚è©¦ç”±éµç›¤è®€å…¥ä¸€å€‹å¸¶æœ‰ä¸€å€‹æˆ–æ•¸å€‹åƒåˆ†ä½ç¬¦è™Ÿçš„æ•´æ•¸ï¼Œç„¶å¾Œå°å‡ºé€™å€‹æ•´æ•¸ä¹˜ä¸Š2ä¹‹å¾Œçš„çµæœã€‚

### 22. æ•´æ•¸ ğ‘› çš„å› æ•¸å°±æ˜¯å¯ä»¥æ•´é™¤ ğ‘› çš„æ•¸ã€‚è©¦çµ¦äºˆä¸€å€‹æ•´æ•¸ï¼Œç„¶å¾Œåˆ—å°å‡ºå®ƒæ‰€æœ‰çš„å› æ•¸ï¼Œä¸¦ç®—å‡ºå› æ•¸çš„å€‹æ•¸ã€‚ä¾‹å¦‚ï¼Œè¼¸å…¥æ•´æ•¸20ï¼Œå‰‡å°å‡º â€œ20çš„å› æ•¸æœ‰6å€‹ï¼Œåˆ†åˆ¥ç‚º1, 2, 4, 5, 10, 20â€ã€‚

```python
n=int(input('è¼¸å…¥ä¸€å€‹æ•´æ•¸n:'))
lst=[]
cnt=1
while cnt<=20:
  if(n%cnt==0):
    lst.append(cnt)
  cnt+=1
res=str(lst)
print(f'{n}çš„å› æ•¸æœ‰{len(lst)}å€‹ï¼Œåˆ†åˆ¥ç‚º{res[1:-1]}')
```

### 23. è³ªæ•¸æ˜¯é™¤äº†1å’Œå®ƒæœ¬èº«ä¹‹å¤–ï¼Œæ²’æœ‰å…¶å®ƒå› æ•¸çš„æ•¸ï¼Œä¹Ÿå°±æ˜¯èªªè³ªæ•¸åªæœƒæœ‰å…©å€‹å› æ•¸ã€‚è«‹åˆ©ç”¨é€™å€‹è§€å¿µä¾†åˆ¤åˆ¥è¼¸å…¥çš„æ•´æ•¸æ˜¯å¦ç‚ºè³ªæ•¸ã€‚

## 5.6 ä¸²åˆ—ç”Ÿæˆå¼ (List comprehension)

### 32. è©¦ä»¥ä¸²åˆ—ç”Ÿæˆå¼å®Œæˆä¸‹åˆ—å„é¡Œï¼š

```python
#(a) å»ºç«‹ä¸€å€‹1åˆ°10çš„å¹³æ–¹æ‰€çµ„æˆçš„ä¸²åˆ—ã€‚
lst=[item**2 for item in range(1,11,1)]
print(lst)
```

```python
#(b) å»ºç«‹ä¸€å€‹1åˆ°50ä¹‹é–“ï¼Œå¯ä»¥åŒæ™‚è¢«3å’Œ4æ•´é™¤çš„ä¸²åˆ—ã€‚
lst=[item for item in range(1,51,1) if item%3==0 and item%4==0]
print(lst)
```

```python
#(c) å–å‡ºå­—ä¸² 'List comprehension' ä¸­æ‰€æœ‰çš„æ¯éŸ³ï¼Œä¸¦çµ„æˆä¸€å€‹å­—å…ƒä¸²åˆ—ã€‚
s='List comprehension'
target='AEIOUaeiou'
lst=[item for item in s if item in target]
print(lst)
```

```python
#(d) ä¾ä¸²åˆ— [3,âˆ’1,4,7,âˆ’3,2] çš„å€¼ä¾†å»ºç«‹å¦ä¸€å€‹æ–°ä¸²åˆ—ï¼Œè‹¥ä¸²åˆ—å…ƒç´ çš„å€¼ç‚ºæ­£ï¼Œå‰‡æ–°ä¸²åˆ—å…ƒç´ çš„å€¼ç‚º1ï¼Œå¦å‰‡ç‚º âˆ’1ã€‚æœ¬é¡Œå»ºç«‹å‡ºä¾†çš„ä¸²åˆ—æ‡‰ç‚º [1,âˆ’1,1,1,âˆ’1,1]ã€‚
lst=[3,-1,4,7,-3,2]
res=[-1 if item <0 else 1 for item in lst]
print(res)
```

### 33. è©¦ä»¥ä¸²åˆ—ç”Ÿæˆå¼å–å‡º ['Spring', 'Summer', 'Autumn', 'Winter'] ä¸­æ‰€æœ‰çš„æ¯éŸ³ï¼Œä¸¦çµ„æˆä¸€å€‹æ–°çš„å­—å…ƒä¸²åˆ—ã€‚

```python
sts=['Spring', 'Summer', 'Autumn', 'Winter']
target='AEIOUaeiou'
res=[ch for st in sts for ch in st if ch in target]
print(res)
```
