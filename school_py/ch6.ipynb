## 2.試設計pow(𝑥,𝑛) 函數，可用來計算並傳回𝑥的𝑛次方。

```python
import math
def pow(x,n):
  if(n==0):
    return 1
  elif(n==1):
    return x
  elif(n%2==1):
    return x*pow(x**2,n//2)
  else:
    return pow(x**2,n//2)
print(pow(2,10))
```

## 3.試設計isprime(𝑥) 函數，用來判別𝑥是否為質數。如果是，則傳回True，否則傳回False。

```python
def isprime(n):
  if(n==2 or n==3):
    return True
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
print(isprime(2))
print(isprime(10))
print(isprime(11))
print(isprime(13))
```

## 4.試設計primes(𝑥) 函數，可以用來找出小於等於𝑥的所有質數，並將找出來的質數以串列的型式傳回。

```python
def primes(x):
  my_list=[]
  if(x>=2):
    my_list.append(2)
  if(x>=3):
    my_list.append(3)
  for i in range(5,x+1,6):
    if(isprime(i)):
      my_list.append(i)
    if(isprime(i+2)):
      my_list.append(i+2)
  return my_list
print(primes(100))
```

## 6.試設計一函數factor(𝑥)，可以將整數參數𝑥進行質因數分解，並將結果以序對組成的串列傳回。例如 72=23×32，因此factor(72) 傳回[(2,3), (3,2)]，又例如330=2 ×3 ×5 ×11，因此factor(330) 傳回[(2,1), (3,1), (5,1), (11,1)]。

```python
def factor(x):
  my_list=[]
  primes_list=primes(x)
  for i in primes_list:
    rd=0
    while x%i==0:
      x/=i
      rd+=1
    if(rd!=0):
      my_list.append((i,rd))
  return my_list
print(factor(72))
print(factor(330))
```

## 7.試設計一函數bin2dec(bs)，用來將二進位的數字bs（為一字串，裡面的字元只能是0或1，bs可以是任意長度）轉換成10進位的整數。如果bs的長度為0，則傳回None。提示：'1011' = 1 ×8+0 ×4+1 ×2+1 ×1=11，因此bin2dec('1011') 傳回11。

```python
def bin2dec(bs):
  if(len(bs)==0):
    return None
  res=0
  m=1
  while len(bs)!=0:
    if(bs[-1]=='1'):
      res+=m
    m*=2
    bs=bs[:-1]
  return res
print(bin2dec('1011'))
```

## 9.試撰寫一函數n_digits(num)，它可接收一個整數num，然後輸出num是幾個位數的整數。例如輸入n_digits(5591)，則輸出4。

```python
def n_digits(num):
  res=0
  while num!=0:
    num//=10
    res+=1
  return res
print(n_digits(5591))
print(n_digits(1))
print(n_digits(123456789))
```

## 10.一個數如果是所有小於它本身的因數之和，這個數就稱為完美數（Perfect  number）。例如6的因數中，小於6的因數有1, 2, 3且6 =1 +2 +3，所以6是完美數。另外，28=1 +2 +4 +7 +14，所以28也是一個完美數。(a)試撰寫一函數factors(𝑛)，可以傳回𝑛的因數中，所有小於𝑛的因數。(b)試利用(a) 的結果找出4個位數以下所有的完美數。

```python
def factors(n):
  sum = 1;
  for i in range(2,n//2+1,1):
    if(n%i==0):
      sum+=i
  return sum
for i in range(10000):
  if(factors(i)==i):
    print(i)
```

## 11.試撰寫一函數square_tup(tpl)，可以用來接收一個內含任意個整數的序對tpl，傳回值為一個序對，內含這些整數的平方。例如square_tup((2,4,6)) 會傳回(4,16,36)。

```python
def square_tup(tpl):
  squared_values = tuple(x ** 2 for x in tpl)
  return squared_values
print(square_tup((2,4,6)))
```

## 15.試設計一函數find_int(num)，可以傳回不能被4整除，且最靠近奇數num的偶數。例如find_int(19)=18，find_int(7)=6，find_int(27)=26。如果num未填，則預設為3。

```python
def find_int(num=3):
  num1=(num//2)*2
  num2=(num//2+1)*2
  if(num1%4==0):
    return num2
  else:
    return num1
print(find_int(19))
print(find_int(7))
print(find_int(27))
print(find_int())
```

## 16.十七世紀的數學家歐拉（Euler）找到了一個計算圓周率的公式（好厲害，不知道他怎麼推導的）：公式, 其中所有的分子都是大於2的質數，分母則是不能被4整除，且最靠近分子的偶數。試撰寫一函數euler_pi(𝑛)，可以利用上面的公式計算圓周率到第𝑛項。例如若𝑛為5，則計算到13/14這一項。如果𝑛省略，則預設到第10項。

```python
import math
def isprime(n):
  if(n==2 or n==3):
    return True
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def find_down(num):
    num1=(num//2)*2
    num2=(num//2+1)*2
    if(num1%4==0):
      return num2
    else:
      return num1
def euler_pi(n=10):
  pi=2
  if(n>=1):
    pi*=(3/2)
    n-=1
  cnt=5
  while n>0:
    if(isprime(cnt)):
      pi*=(cnt/find_down(cnt))
      #print(f'{cnt} {find_down(cnt)}')
      n-=1
    if(isprime(cnt+2)):
      pi*=((cnt+2)/find_down(cnt+2))
      #print(f'{cnt+2} {find_down(cnt+2)}')
      n-=1
    cnt+=6
  return pi
print(euler_pi(10))
print(euler_pi())
```

## 17.試設計一函數make_dict(key,value) 函數，可以依據給定的key和value來建立並傳回一個字典，其中key和value是長度相等的串列。若value省略，則所有key對應的value皆為0。例如make_dict([0,  1,  2],  [32,  43,  55])  可傳回{0:  32,  1:  43,  2:  55}，而make_dict([0, 1, 2]) 則傳回{0: 0, 1: 0, 2: 0}。

```python
my_dict=dict()
def make_dict(keys,values=None):
  for i in range(len(keys)):
    if values!=None and i<len(values):
      my_dict[keys[i]]=values[i]
    else:
      my_dict[keys[i]]=0
make_dict([0, 1, 2], [32, 43, 55])
print(my_dict)
my_dict.clear()
make_dict([0, 1, 2])
print(my_dict)
```

## 18.試設計一函數to_upper(str1)，它可接收 一個字串str1，傳回值為小寫字母變成大寫之後的字串（其它不是小寫字母的字元不變）；如果str1 省略不傳，則to_upper() 傳回'Null'字串。例如to_upper('Julia_program') 會傳回'JULIA_PROGRAM'。

```python
def to_upper(str1=None):
  if(str1==None):
    return 'Null'
  else:
    return str1.upper()
print(to_upper())
print(to_upper('Julia_program'))
```
