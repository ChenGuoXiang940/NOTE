```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(12,6))
#(a)
ax1 = fig.add_subplot(2, 2, 1)
x = np.linspace(-7, 4, 256)
y = x**4+6*x**3+7*x+3
ax1.plot(x, y)
#(b)
ax2 = fig.add_subplot(2, 2, 2)
y = 6*np.sin(x+3)*np.cos(x)
x = np.linspace(-np.pi, 2*np.pi, 256)
ax2.plot(x, y)
#(c)
ax3 = fig.add_subplot(2, 2, 3)
x = np.linspace(-3, 6, 256)
y = (x+3)/(x**2+1)
ax3.plot(x, y)
#display
plt.show()
```

## 10.設𝑓(𝑛)為小於等於𝑛之質數的個數（𝑛為整數），例如2, 3, 5, 7,11, 13, 17, 19皆為質數，因此𝑓(2)=1，𝑓(7)=4，𝑓(19)=𝑓(20)=8。試繪出𝑓(𝑛)，2≤𝑛≤100的圖形。在此題中，因為𝑛為整數時，𝑓(𝑛)才有定義，所以點和點之間不要以線條連接。

```python
import numpy as np
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
x = np.arange(2, 101)
y = [sum(is_prime(i) for i in range(2, n+1)) for n in x]
plt.scatter(x, y)#散佈圖
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Number of primes less than or equal to n')
plt.show()
```

## 18.試繪出下列各函數的二維、三維等高線圖與3維的曲面圖，並將它們顯示在㇐個 1×3 的子圖中。二維的等高線需顯示等高線的值，figsize參數的值設為  (10, 3)。

### (a)sin(𝑥) sin (𝑦),0≤𝑥≤2𝜋,0≤𝑦≤2𝜋，等高線為 –0.7,−0.3,0,0.3,0.7。

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,3))
x=np.linspace(0,2*np.pi,256)
y=np.linspace(0,2*np.pi,256)
l=[-0.7,-0.3,0,0.3,0.7]
x,y=np.meshgrid(x,y)
z=np.sin(x)*np.sin(y)
# 2D Contour plot
ax1=fig.add_subplot(131)
count=ax1.contour(x,y,z,levels=l)
ax1.clabel(count,inline=True,fontsize=10)
ax1.set_title('2D Contour plot')
# 3D Contour plot
ax2=fig.add_subplot(132,projection='3d')
ax2.contour(x,y,z,levels=l)
ax2.set_title('3D Contour plot')
# 3D Surface plot
ax3=fig.add_subplot(133,projection='3d')
ax3.plot_surface(x,y,z,cmap='viridis')
ax3.set_title('3D Surface plot')
plt.show()
```

```python
fig=plt.figure(figsize=(10,3))
x=np.linspace(-np.pi,np.pi,256)
y=np.linspace(-np.pi,np.pi,256)
x,y=np.meshgrid(x,y)
z=1/np.sqrt(x**2+y**2+1)
# 2D Contour plot
ax1=fig.add_subplot(131)
count=ax1.contour(x,y,z,levels=5)
ax1.clabel(count,inline=True,fontsize=10)
ax1.set_title('2D Contour plot')
# 3D Contour plot
ax2=fig.add_subplot(132,projection='3d')
ax2.contour(x,y,z,levels=5)
ax2.set_title('3D Contour plot')
# 3D Surface plot
ax3=fig.add_subplot(133,projection='3d')
ax3.plot_surface(x,y,z,cmap='viridis')
ax3.set_title('3D Surface plot')
plt.show()
```

## 11.6動畫的製作

## 22.試製作㇐個紅球（大小為10，markersize=10）在二維函數𝑓(𝑥)=tan(sin(𝑥))−sin(tan(𝑥))上方移動的動畫，0≤𝑥≤𝜋，幀數為240，每幀播放的間隔為0.05秒。下圖為其中的三幀動畫：

```python
%matplotlib widget
import matplotlib.animation as animation
fig, ax = plt.subplots()
x = np.linspace(0, np.pi, 240)
f = np.tan(np.sin(x)) - np.sin(np.tan(x))
ax.plot(x, f)
ball, = ax.plot([], [], 'ro', markersize=10)
def update(frame):
    ball.set_data(x[frame], f[frame])
    return ball,
ani = animation.FuncAnimation(fig, update, frames=range(240), interval=50, blit=True)
plt.show()
```

## 23.試製作㇐個紅球（大小為10，markersize=10）在極坐標函數𝑟=sin(6𝑡),0≤𝑡≤2𝜋上方移動的軌跡，移動時𝑟=sin(6𝑡)的圖形會㇐起繪出。幀數取256，每幀播放間隔為0.05秒。下圖為其中三幀動畫（提示：可利用set_data(𝑡,𝑟) 來設定𝑡和𝑟=sin(6𝑡)的坐標）：


```python
fig = plt.figure()
t = np.linspace(0, 2*np.pi, 256)
r = np.sin(6*t)
ax = fig.add_subplot(111, polar=True)
ax.plot(t, r)
ball, = ax.plot([], [], 'ro', markersize=10)
def update(frame):
    ball.set_data(t[frame], r[frame])
    return ball,
ani = animation.FuncAnimation(fig, update, frames=range(256), interval=50, blit=True)
plt.show()
```
