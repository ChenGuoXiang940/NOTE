## 22.è©¦æ’°å¯«éè¿´å‡½æ•¸r_pow(base,n)ï¼Œç”¨ä¾†è¨ˆç®—baseçš„næ¬¡æ–¹ï¼Œä¸¦åˆ©ç”¨æ­¤å‡½æ•¸ä¾†è¨ˆç®— 2^8ã€‚

```python
def r_pow(base,n):
  if(n==0):
    return 1
  elif(n==1):
    return base
  elif((base%2)==1):
    return base*r_pow(base**2,n//2)
  else:
    return r_pow(base**2,n//2)
print(r_pow(2,8))
print(r_pow(10,5))
```

## 32.è²»å¼æ•¸åˆ—ä»»ä¸€é …çš„å€¼ç­‰æ–¼å‰å…©é …ä¹‹å’Œï¼Œä¸”ç¬¬1é …å’Œç¬¬2é …çš„å€¼å‡ç‚º1ï¼Œä¾‹å¦‚è²»æ°æ•¸åˆ—çš„å‰10é …åˆ†åˆ¥ç‚º1, 1, 2, 3, 5, 8, 13, 21, 34å’Œ55ã€‚è©¦ä¾åºä½œç­”ä¸‹åˆ—å„é¡Œï¼š

### (a)è©¦å¯«ä¸€å‡½æ•¸fib(ğ‘›)ï¼Œå¯ä»¥å‚³å›å‰ğ‘›å€‹è²»å¼æ•¸åˆ—çš„å€¼ã€‚

```python
def fib(n):
  lst=[]
  if(n>=1):
    lst.append(1)
  if(n>=2):
    lst.append(1)
  a,b=1,1
  for i in range(2,n,1):
    c=a+b
    lst.append(c)
    a=b
    b=c
  return lst
print(fib(10))
```

### (b)è©¦å°‡(a) æ”¹å¯«æˆä¸€å€‹è²»å¼æ•¸åˆ—ç”¢ç”Ÿå™¨fib_gen(ğ‘›)ã€‚

```python
def fib_gen():
  yield 1
  yield 1
  a,b=1,1
  while True:
    c=a+b
    yield c
    a=b
    b=c
gi=fib_gen()
for i in range(10):
  print(next(gi))
```

## 33.è©¦æ’°å¯«ä¸€å€‹ç”¢ç”Ÿå™¨gen(ğ‘›)ï¼Œå¯ä»¥ç”¢ç”Ÿæ‰€æœ‰å°æ–¼ç­‰æ–¼ğ‘›çš„æ•´æ•¸ä¸­  ï¼Œå¯ä»¥è¢«3æ•´é™¤ï¼Œä½†ä¸èƒ½è¢«5æ•´é™¤çš„æ•¸ã€‚

## 34.è©¦æ’°å¯«ä¸€å€‹è³ªæ•¸ç”¢ç”Ÿå™¨primes(ğ‘›)ï¼Œå¯ä»¥ç”¨ä¾†ç”¢ç”Ÿå°æ–¼ç­‰æ–¼ğ‘›çš„è³ªæ•¸ã€‚

```python
import math
def isprime(n):
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq+1,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def primes(n):
  lst=[]
  if(n>=2):
    lst.append(2)
  if(n>=3):
    lst.append(3)
  cnt=5
  while cnt<=n:
    if(isprime(cnt)==True):
      lst.append(cnt)
    if(isprime(cnt+2)==True):
      lst.append(cnt+2)
    cnt+=6
  return lst
print(primes(100))
```

## 5.è©¦æ’°å¯«ä¸€å€‹è³ªæ•¸ç”¢ç”Ÿå™¨primes()ï¼Œå¯ä»¥ç”¨ä¾†ç”¢ç”Ÿç„¡çª®å¤šå€‹è³ªæ•¸ã€‚äº¦å³åªè¦æŠŠprimes() å»ºç«‹çš„ç‰©ä»¶ğ‘”å‚³å…¥next() å‡½æ•¸ï¼Œå°±å¯ä»¥å‚³å›ä¸‹ä¸€å€‹è³ªæ•¸ã€‚

```python
import math
def isprime(n):
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq+1,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def primes_gen():
  yield 2
  yield 3
  cnt=5
  while True:
    if(isprime(cnt)==True):
      yield cnt
    if(isprime(cnt+2)==True):
      yield(cnt+2)
    cnt+=6
gi=primes_gen()
for i in range(10):
  print(next(gi))
```
