，如果把列和行的索引看成是坐標 (𝑟, 𝑐)，試作答下列各題

### (a) 試找出離坐標 (2,2) 最近，且元素值為 4 的坐標（答案為 (1,2)）。

```python
a=np.array([[0,1,2,3,4],[2,3,4,0,1],[4,0,1,2,3],[1,2,3,4,0],[3,4,0,1,2]])
def getpoint(target,item):
    indexs=np.where(a==item)
    dist=np.sqrt((indexs[0] - target[0]) ** 2 + (indexs[1] - target[1]) ** 2)
    min_dist_index= np.argmin(dist)
    return (indexs[0][min_dist_index], indexs[1][min_dist_index])
print(getpoint((2,2),4))
```

### (b) 試找出離坐標 (4,3) 最近，且元素值為 0 的坐標（答案為 (4,2)）。

```python
print(getpoint((4,0),2))
```

### (d) 找出相距最遠的兩個 4 的坐標（答案應為 (0,4) 和 (4,1)）。

```python
indexs=np.argwhere(a==4)
#print(indexs)
dist = np.sqrt(np.sum((indexs[:, None] - indexs) ** 2, axis=-1))
#print(dist)
i, j = np.unravel_index(np.argmax(dist), dist.shape)
print(indexs[i],indexs[j])
```

## 6. 試設計一函數 prime_list(𝑛)，可以找出小於 𝑛 的所有質數

```python
def isprime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
def prime_list(n):
    return [n for n in np.arange(1, n+1) if isprime(n)]
print(prime_list(100))
```
