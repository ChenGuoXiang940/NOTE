## 1.è¨­Windowé¡åˆ¥å®šå¦‚ä¸‹ï¼Œè©¦å›ç­”ä¸‹åˆ—å•é¡Œï¼š

```python
class Window:
  def __init__(self,w=10,h=5):
    self.width=w
    self.height=h
  def area(self):
    return self.width*self.height
  def cmp(self,other):
    return max(self.area(),other.area())
#(1)è©¦å»ºç«‹ä¸€å€‹Windowé¡åˆ¥çš„ç‰©ä»¶ğ‘¤0ï¼Œå…¶å±¬æ€§widthèˆ‡heightçš„å€¼ç‚ºé è¨­å€¼
w0=Window()
#(2)å»ºç«‹ä¸€å€‹Windowé¡åˆ¥çš„ç‰©ä»¶ğ‘¤1ï¼Œä¸¦åˆ†åˆ¥è¨­å®šå±¬æ€§widthèˆ‡heightçš„å€¼ç‚º12å’Œ8ã€‚
w1=Window(12,8)
#(3)è©¦åˆ©ç”¨setattr() åˆ†åˆ¥å°‡ğ‘¤1 çš„å±¬æ€§widthèˆ‡heightæ”¹è¨­ç‚º16å’Œ7ã€‚
setattr(w1,'width',16)
setattr(w1,'height',7)
print(w1.area())
#(4)è©¦æ¯”è¼ƒğ‘¤0å’Œğ‘¤1 çš„é¢ç©å¤§å°ï¼ˆwidth x heightï¼‰ï¼Œ ç„¶å¾Œå°å‡ºè¼ƒå¤§çš„é¢ç©ã€‚
print(w0.cmp(w1))
```

## 4.å·²çŸ¥çƒé«”ç©ç‚º4/3*ğœ‹*r^3ï¼Œè¡¨é¢ç©ç‚º4ğœ‹*r^2ï¼Œè©¦å®Œæˆä¸‹åˆ—å„é¡Œï¼š

```python
import math
class Sphere:
  def __init__(self,r):
    self.rad=r
  def volume(self):
    return 4/3*math.pi*self.rad**3
  def surface_area(self):
    return 4*math.pi*self.rad**2
  def __repr__(self):
    return f"Sphere object, rad ={self.rad}"
  def __str__(self):
    return f"sphere object,rad={self.rad},volumn={round(self.volume(),2)},surface_area={round(self.surface_area(),2)}"
#(a)è©¦å»ºç«‹ä¸€å€‹Sphereçš„é¡åˆ¥ï¼Œå…§å«ä¸€å€‹init(r) å‡½æ•¸ï¼Œå¯å°‡radå±¬æ€§è¨­å€¼ç‚ºrã€‚
s0=Sphere(2)
#(b)è©¦åœ¨Sphereé¡åˆ¥è£¡å®šç¾©volume()å‡½æ•¸ï¼Œå¯ç”¨ä¾†å‚³å›åœ“çƒçš„é«”ç©ï¼Œèˆ‡ä¸€å€‹surface_area() å‡½æ•¸ï¼Œå¯ä»¥å‚³å›åœ“çƒçš„è¡¨é¢ç©
#(c)è©¦å»ºç«‹ä¸€å€‹rad=2çš„Sphereé¡åˆ¥ä¹‹ç‰©ä»¶ğ‘ 0ï¼Œä¸¦æ±‚å‡ºæ­¤ç‰©ä»¶çš„é«”ç©å’Œè¡¨é¢ç©
print(f'{s0.volume()} {s0.surface_area()}')

```

## 6.è©¦è¨­è¨ˆä¸€å€‹Calculatoré¡åˆ¥ï¼Œå…§å«__init__(a,b) å‡½æ•¸ï¼Œå¯å°‡n1èˆ‡n2å±¬æ€§åˆ†åˆ¥è¨­ç‚ºaèˆ‡bï¼Œä¸¦è¨­è¨ˆä¸‹åˆ—å„å‡½æ•¸ï¼Œç„¶å¾Œä»¥ç‰©ä»¶c0= Calculator(2,10) ä¾†æ¸¬è©¦å®ƒå€‘:

```python
import math
class Calculator:
  def __init__(self,a,b):
    self.n1=a
    self.n2=b
  def add(self):
    return self.n1+self.n2
  def gcd(self):
    return math.gcd(self.n1,self.n2)
  def lcm(self):
    return (self.n1/self.gcd())*self.n2
  def power(self):
    return (self.n1**2,self.n2**2)
obj=Calculator(2,10)
#(a)è¨­è¨ˆadd()å‡½æ•¸ï¼Œå¯å‚³å›n1èˆ‡ğ‘›2ä¹‹å’Œã€‚
print(obj.add())
#(b)è¨­è¨ˆgcd()å‡½æ•¸ï¼Œå¯å‚³å›n1èˆ‡n2çš„æœ€å¤§å…¬å› æ•¸
print(obj.gcd())
#(c)è¨­è¨ˆlcm()å‡½æ•¸ï¼Œå¯å‚³å›n1èˆ‡n2çš„æœ€å°å…¬å€æ•¸ã€‚
print(obj.lcm())
#(d)è¨­è¨ˆpower()å‡½æ•¸ï¼Œå¯å‚³å›n1èˆ‡n2æ¬¡æ–¹ã€‚
print(obj.power())
```
