## 2. è©¦ä¾åºå®Œæˆä¸‹åˆ—å„é¡Œï¼š

### (a) è©¦æ’°å¯«ä¸€ç¨‹å¼æ‰¾å‡º 100 çš„æ‰€æœ‰å› æ•¸ï¼Œä¸¦å°‡æ‰¾å‡ºçš„å› æ•¸å¯«å…¥ç´”æ–‡å­—æª” factors.txt ä¸­ï¼Œæ¯å€‹å› æ•¸ç”¨é€—è™Ÿéš”é–‹ã€‚

```python
f=open('factors.txt','w',encoding='utf-8')
lst=[]
for i in range(1,101,1):
  if(100%i==0):
    lst.append(i)
f.write(f"{' '.join(map(str,lst))}")
f.close()
```

### (b) é–‹å•Ÿ factors.txtï¼Œè®€å–å¯«å…¥çš„å› æ•¸ï¼Œç„¶å¾Œè¨ˆç®—å…¶ç¸½å’Œã€‚

```python
f=open('factors.txt','r',encoding='utf-8')
lst=[eval(i) for i in f.readline().replace(","," ").split()]
f.close()
print(sum(lst))
```

### (c) ä»¥äºŒé€²ä½æª”çš„æ ¼å¼å°‡ 100 çš„æ‰€æœ‰å› æ•¸å¯«å…¥ factors.bin ä¸­ã€‚

```python
with open('factors.bin','wb') as f:
  lst=[]
  for i in range(1,101,1):
    if(100%i==0):
      lst.append(i)
  f.write(bytes(lst))
  f.close()
```

### (d) è®€å–å„²å­˜æ–¼ factors.bin ä¸­çš„æ•¸å­—ï¼Œç„¶å¾Œè¨ˆç®—å…¶ç¸½å’Œã€‚

```python
with open('factors.bin','rb') as f:
  lst=[i for i in f.read()]
  f.close()
  print(sum(lst))
```

## 3. è¨­ä¸€å€‹ç´”æ–‡å­—æª” mat.txt çš„å…§å®¹å¦‚ä¸‹ï¼ˆæ•¸å­—ä¹‹é–“æœ‰ä¸€å€‹ç©ºæ ¼ï¼‰ï¼Œè©¦è®€å–é€™å€‹ç´”æ–‡å­—æª”ï¼Œç„¶å¾Œæ‰¾å‡ºè£¡é¢æ•¸å­—çš„æœ€å¤§å€¼å’Œæœ€å°å€¼ã€‚


```python
with open('mat.txt','r') as f:
  val_max=-1E5
  val_min=1E5
  for line in f.readlines():
    for i in line.split():
      val_max=max(val_max,eval(i))
      val_min=min(val_min,eval(i))
  print(f'æœ€å¤§å€¼:{val_max} æœ€å°å€¼:{val_min}')
```

## 5. è©¦å®Œæˆä¸‹åˆ—å„é¡Œï¼š

### (a) æ‰¾å‡ºå°æ–¼ 1000 çš„æ‰€æœ‰è³ªæ•¸ï¼Œä¸¦ä»¥ä¸€å€‹ä¸²åˆ— lst å­˜æ”¾ã€‚

```python
isprime=[True]*1000
for i in range(2,1000,1):
  if(isprime[i]):
    for j in range(i+i,1000,i):
      isprime[j]=False
isprime[0]=isprime[1]=False
lst=[]
for i in range(0,1000,1):
  if(isprime[i]):
    lst.append(i)
print(lst)
```
