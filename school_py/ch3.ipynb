### 16.設𝑎=16，𝑏=21.5639，試利用format() 函數將𝑎和𝑏的值進行如下的輸出，其中符號 ◌ 代表一個空格：

```python
import math
𝑎=16
𝑏=21.5639
print('a={:8.2f}, b={:8.3f}'.format(a,b))
print('a={:8d}, b={:08.3f}'.format(a,b))
print('a={:<+8.1f}, b={:<+8.3f}'.format(a,b))
print('a={:>8}, b={:>8d}'.format('0x{:x}'.format(a),math.ceil(b)))
print('a={:>8}, b={:+08.2f}'.format('0o{:o}'.format(a),b))
print('a={{{:>6}}}, b={{{:>6.2f}}}'.format('0o{:o}'.format(a),b))
print('a={:>8}, b={:<08.5f}'.format('\''+'0o{:o}'.format(a)+'\'',b))
print('a={{\'{:4}\'}}, b=\'{{{:>5.2f}}}\''.format('0o{:o}'.format(a),b))
```
