```python
import numpy.random as random
import numpy as np
import matplotlib.pyplot as plt
Gaussian_noise=random.uniform(-1,1,1000)
t=np.linspace(0,1,1000,endpoint=False)
x1=3*np.cos(2*np.pi*10*t)
x2=x1+Gaussian_noise
figure,axis=plt.subplots(3)
axis[0].plot(t,x1,'-',label='Sinusoidal(t)')
axis[0].legend(loc='upper right')
axis[0].set_ylabel('Amplitude')
axis[0].axis([0,1,-6,6])

axis[1].plot(t,Gaussian_noise,'-',label='Noise(t)')
axis[1].legend(loc='upper right')
axis[1].set_ylabel('Amplitude')
axis[1].axis([0,1,-3,3])

axis[2].plot(t,x2,'-',label='x2(t)')
axis[2].legend(loc='upper right')
axis[2].set_ylabel('Amplitude')
axis[2].axis([0,1,-6,6])
plt.show()
```

```python
import numpy.random as random
import numpy as np
import matplotlib.pyplot as plt
n1=random.uniform(-1,1,1000)
ns=np.cumsum(n1)
mean=np.mean(ns)
max1=np.max(np.absolute(ns-mean))
Brownian_boise=(ns-mean)/max1
t=np.linspace(0,1,1000,endpoint=False)
x1=3*np.cos(2*np.pi*10*t)
x2=x1+Brownian_boise
figure,axis=plt.subplots(3)
axis[0].plot(t,x1,'-',label='Sinusoidal(t)')
axis[0].legend(loc='upper right')
axis[0].set_ylabel('Amplitude')
axis[0].axis([0,1,-6,6])

axis[1].plot(t,Brownian_boise,'-',label='Noise(t)')
axis[1].legend(loc='upper right')
axis[1].set_ylabel('Amplitude')
axis[1].axis([0,1,-3,3])

axis[2].plot(t,x2,'-',label='x2(t)')
axis[2].legend(loc='upper right')
axis[2].set_ylabel('Amplitude')
axis[2].axis([0,1,-6,6])
plt.show()
```
