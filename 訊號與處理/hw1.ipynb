```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,1,1000,endpoint=False)
x=1*np.cos(2*np.pi*2*t)

plt.plot(t,x)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')

plt.show()
```
