```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,4,3,2,1,1])
h=np.array([1,2,3,1,1])
y=np.convolve(x,h,'full')
y1=np.convolve(x,h,'same')
y2=np.convolve(x,h,'valid')
print('x=',x)
print('full convolution',y)
print('convolution y=',y1)
plt.figure(1)
plt.stem(y)
```
