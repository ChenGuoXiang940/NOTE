```python
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,1,1000,endpoint=False)
x1=3*np.cos(2+np.pi*10*t+np.pi/4)
x2=2*np.cos(2*np.pi*10*t+np.pi/4)
x3=x1+x2
plt.rcParams["figure.figsize"]=(10,8)
plt.plot(t,x1,'-',label='x1(t)')
plt.plot(t,x2,'--',label='x2(t)')
plt.plot(t,x3,'-',label='x3(t)')
plt.legend(loc='upper right')
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([0,1,-6,6])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,1,1000,endpoint=False)
x1=4*np.cos(2+np.pi*10*t+np.pi/3)
x2=3*np.cos(2*np.pi*10*t+5*np.pi/6)
x3=x1+x2
plt.rcParams["figure.figsize"]=(10,8)
plt.plot(t,x1,'-',label='x1(t)')
plt.plot(t,x2,'--',label='x2(t)')
plt.plot(t,x3,'-',label='x3(t)')
plt.legend(loc='upper right')
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([0,1,-10,10])
plt.show()

phasor_x1 = 4 * np.exp(1j * (2 + np.pi / 3))
phasor_x2 = 3 * np.exp(1j * (5 * np.pi / 6))
phasor_x3 = phasor_x1 + phasor_x2

print('Phasor x1:', phasor_x1)
print('Phasor x2:', phasor_x2)
print('Phasor x3:', phasor_x3)

A_x3 = np.abs(phasor_x3)
phi_x3 = np.angle(phasor_x3)

print('Amplitude A of x3(t):', A_x3)
print('Phase shift  of x3(t):', phi_x3)
```
