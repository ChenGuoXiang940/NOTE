##21.設total(𝑛)   =1 +2 +3 +⋯+𝑛，試利用遞迴的方式來撰寫函數total(𝑛)（提示：total(𝑛) = 𝑛 + total(𝑛−1)   且total(1)=1。

```python
def total(n):
  if(n==1):
    return 1
  return n+total(n-1)
print(total(4))
print(total(10))
```

##22.試撰寫遞迴函數r_pow(base,n)，用來計算base的n次方，並利用此函數來計算 2^8。

```python
def r_pow(base,n):
  if(n==0):
    return 1
  elif(n==1):
    return base
  elif((base%2)==1):
    return base*r_pow(base**2,n//2)
  else:
    return r_pow(base**2,n//2)
print(r_pow(2,8))
print(r_pow(10,5))
```

##23.試撰寫遞迴函數r_sum(𝑛 ) 來求算1 ×2 +2 ×3 +3 ×4 +⋯+(𝑛−1)×𝑛之和，並以𝑛=5 來測試您的結果。

```python
def r_sum(n):
  if(n==2):
    return 2
  return (n-1)*n+r_sum(n-1)
print(r_sum(5))
print(r_sum(3))
```

##32.費式數列任一項的值等於前兩項之和，且第1項和第2項的值均為1，例如費氏數列的前10項分別為1, 1, 2, 3, 5, 8, 13, 21, 34和55。試依序作答下列各題：

###(a)試寫一函數fib(𝑛)，可以傳回前𝑛個費式數列的值。

```python
def fib(n):
  lst=[]
  if(n>=1):
    lst.append(1)
  if(n>=2):
    lst.append(1)
  a,b=1,1
  for i in range(2,n,1):
    c=a+b
    lst.append(c)
    a=b
    b=c
  return lst
print(fib(10))
```

###(b)試將(a) 改寫成一個費式數列產生器fib_gen(𝑛)。

```python
def fib_gen():
  yield 1
  yield 1
  a,b=1,1
  while True:
    c=a+b
    yield c
    a=b
    b=c
gi=fib_gen()
for i in range(10):
  print(next(gi))
```

##33.試撰寫一個產生器gen(𝑛)，可以產生所有小於等於𝑛的整數中  ，可以被3整除，但不能被5整除的數。

```python
def gen(n):
  cnt=3
  while cnt<=n:
    if(cnt%5!=0):
      yield cnt
    cnt+=3
gi=gen(100)
print(list(gi))
```

##34.試撰寫一個質數產生器primes(𝑛)，可以用來產生小於等於𝑛的質數。

```python
import math
def isprime(n):
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq+1,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def primes(n):
  lst=[]
  if(n>=2):
    lst.append(2)
  if(n>=3):
    lst.append(3)
  cnt=5
  while cnt<=n:
    if(isprime(cnt)==True):
      lst.append(cnt)
    if(isprime(cnt+2)==True):
      lst.append(cnt+2)
    cnt+=6
  return lst
print(primes(100))
```

##5.試撰寫一個質數產生器primes()，可以用來產生無窮多個質數。亦即只要把primes() 建立的物件𝑔傳入next() 函數，就可以傳回下一個質數。

```python
import math
def isprime(n):
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq+1,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def primes_gen():
  yield 2
  yield 3
  cnt=5
  while True:
    if(isprime(cnt)==True):
      yield cnt
    if(isprime(cnt+2)==True):
      yield(cnt+2)
    cnt+=6
gi=primes_gen()
for i in range(10):
  print(next(gi))
```
