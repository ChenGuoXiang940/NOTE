## 4.設𝑙s = [9, 8, 7, 1, 2, 3, 7, 3, 2]，試完成下列各題：

```python
str1 = [9, 8, 7, 1, 2, 3, 7, 3, 2]
#(a)取出𝑙𝑙𝑠𝑠𝑠𝑠中，索引為0到2的元素（含索引2）。
print(str1[:3])
#(b)取出𝑙𝑙𝑠𝑠𝑠𝑠中最後3個元素。
print(str1[-3:])
#(c)取出𝑙𝑙𝑠𝑠𝑠𝑠中，索引為4到7的元素（含索引4）。
print(str1[4:7])
#(d)取出𝑙𝑙𝑠𝑠𝑠𝑠中，索引為偶數的元素（不包含索引為0的元素）
print(str1[0::2])
#(e)找出𝑙𝑙𝑠𝑠𝑠𝑠的長度、最大值與最小值，並計算𝑙𝑙𝑠𝑠𝑠𝑠元素的總和。
print(f"長度:{len(str1)} 最大值:{max(str1)} 最小值:{min(str1)} 元素總和:{sum(str1)}")
#(f)反向提取  倒數第1個到倒數第4個元素，即提取結果應為[2, 3, 7, 3]
print(str1[-1:-5:-1])
#(g)將𝑙s反向排序，結果應為  [2, 3, 7, 3, 2, 1, 7, 8, 9]。
print(str1[-1::-1])
```

## 14.於下列各小題中皆假設𝑠1 ={1, 2, 4, 4, 8}和𝑠2 ={4, 5, 6, 7, 8}，試完成下列各題：

```python
𝑠1 ={1, 2, 4, 4, 8}
𝑠2 ={4, 5, 6, 7, 8}
#(a)將元素5添加到𝑠1 中。
s1.add(5)
#(b)刪除掉𝑠1 中的元素8。
s1.discard(8)
#(c)找出𝑠1 有，但𝑠2 沒有的元素。這個結果是否會與𝑠2 有，但𝑠1 沒有的元素相同?
print(s1.difference(s2))
#(d)找出𝑠1 和𝑠2 共有的元素（即交集）。
print(s1.intersection(s2))
#(e)找出𝑠1 和𝑠2 不是共有的元素。
print(s1.symmetric_difference(s2))
#(f)求出𝑠1 和𝑠2 的聯集。
print(s1.union(s2))
#(g)用𝑠2 的值來更新𝑠1（𝑠2 有，但𝑠1 沒有的元素會加入𝑠1 中）。
s1.update(s2)
print(s1)
#(h)刪除𝑠1 和𝑠2 的所有元素
s1.clear()
s2.clear()
print(f'{s1} {s2}')
```
