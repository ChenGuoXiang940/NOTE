### (e) å–å‡ºé™£åˆ— ğ‘ ä¸­ï¼Œæ‰€æœ‰å¯ä»¥è¢« 3 æ•´é™¤çš„æ•¸ã€‚

```python
b=a[a%3==0]
print(b)
```

### (f) æ‰¾å‡ºé™£åˆ— ğ‘ ä¸­ï¼Œåˆ—èˆ‡è¡Œçš„ç´¢å¼•ä¹‹å’Œç‚ºå¶æ•¸çš„æ‰€æœ‰å…ƒç´ ï¼Œä¸¦å°‡å®ƒå€‘æ”¾åœ¨ä¸€å€‹ä¸²åˆ—è£¡ï¼ˆæœ¬é¡Œå¯èƒ½éœ€è¦æ’°å¯«è¿´åœˆï¼‰ã€‚

```python
lst=a.tolist()
ans=[]
m=0
n=0
for i in lst:
  for j in i:
    if((m+n)%2==0):
      ans.append(j)
    n+=1
  m+=1
  n=0
print(ans)
```

### (g) æ‰¾å‡ºé™£åˆ— ğ‘ ä¸­ï¼Œå¯ä»¥è¢« 2 æ•´é™¤ï¼Œä½†ä¸èƒ½è¢« 3 æ•´é™¤çš„å…ƒç´ å…±æœ‰å¹¾å€‹ã€‚

```python
ans=0
for i in a.tolist():
  for j in i:
    if(j%2==0 and j%3!=0):
      ans+=1
print(ans)
```
