### 2.試輸入下面不同進位的數字，並將它們轉換成十進位：(a)16進位的數字a0ff、121ab、acd123。(b)8進位的數字7231、754、567。(c)2進位的數字11001010、1001001、110101000100010110。

```python
print(int('a0ff',16))  #print(0xa0ff)
print(int('121ab',16)) #print(0x121ab)
print(int('acd123',16))#print(0xacd123)
```

```python
print(int('7231',8))#print(0o7231)
print(int('754',8)) #print(0o754)
print(int('567',8)) #print(0o567)
```

```python
print(int('11001010',2))          #print(0b11001010)
print(int('1001001',2))           #print(0b1001001)
print(int('110101000100010110',2))#print(0b110101000100010110)
```

### 8.試思考下列各式的運算結果，並利用Python驗證您思考的結果是否正確。

```python
print(8<3 and 7==9 and 7>4)     #8<3是錯的,7==9是錯的7>4是對的,全部做和邏輯遇 Flase 則為 False
print(2>8 or 7<=8 and 7>2)      #2>8是錯的,7<=8是對的,7>2是對的,and 優先權較 or 高,先做和邏輯得True,再來做或邏輯得True
print(False and 'kitty' or True)#False代表錯的,字元是對的,True是對的,and 優先權較 or 高,先做和邏輯得False,再來做或邏輯得True
```
