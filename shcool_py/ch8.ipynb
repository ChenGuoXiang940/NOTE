## 2. 試依序完成下列各題：

### (a) 試撰寫一程式找出 100 的所有因數，並將找出的因數寫入純文字檔 factors.txt 中，每個因數用逗號隔開。

```python
f=open('factors.txt','w',encoding='utf-8')
lst=[]
for i in range(1,101,1):
  if(100%i==0):
    lst.append(i)
f.write(f"{' '.join(map(str,lst))}")
f.close()
```

### (b) 開啟 factors.txt，讀取寫入的因數，然後計算其總和。

```python
f=open('factors.txt','r',encoding='utf-8')
lst=[eval(i) for i in f.readline().replace(","," ").split()]
f.close()
print(sum(lst))
```

### (c) 以二進位檔的格式將 100 的所有因數寫入 factors.bin 中。

```python
with open('factors.bin','wb') as f:
  lst=[]
  for i in range(1,101,1):
    if(100%i==0):
      lst.append(i)
  f.write(bytes(lst))
  f.close()
```

### (d) 讀取儲存於 factors.bin 中的數字，然後計算其總和。

```python
with open('factors.bin','rb') as f:
  lst=[i for i in f.read()]
  f.close()
  print(sum(lst))
```

## 3. 設一個純文字檔 mat.txt 的內容如下（數字之間有一個空格），試讀取這個純文字檔，然後找出裡面數字的最大值和最小值。


```python
with open('mat.txt','r') as f:
  val_max=-1E5
  val_min=1E5
  for line in f.readlines():
    for i in line.split():
      val_max=max(val_max,eval(i))
      val_min=min(val_min,eval(i))
  print(f'最大值:{val_max} 最小值:{val_min}')
```

### (a) 找出小於 1000 的所有質數，並以一個串列 lst 存放。

```python
isprime=[True]*1000
for i in range(2,1000,1):
  if(isprime[i]):
    for j in range(i+i,1000,i):
      isprime[j]=False
isprime[0]=isprime[1]=False
lst=[]
for i in range(0,1000,1):
  if(isprime[i]):
    lst.append(i)
print(lst)
```

### (b) 將 (a) 中的 lst 利用 pickle 模組裡的 dumps() 函數轉成 bytes 物件，然後將它存在primes.bin 檔案中。

### (a) 試利用 StringIO() 將下列的資料寫入記憶體中，然後從記憶體讀取這些數據，並找出它們的最大值。

```python
import csv
from io import StringIO
csv_data='75,87,86,19,69\n78,65,12,77,90'
f0=StringIO(csv_data)
val_max=0
for i in csv.reader(f0,delimiter=','):
  for j in i:
    val_max=max(val_max,eval(j))
print(val_max)
```

```python
import pickle
from io import BytesIO
f1=BytesIO(pickle.dumps([{3,5},'Python',(64,1024)]))
print(pickle.loads(f1.readline()))
```

## 9. 試設計一個程式，在 try 區塊裡由使用者輸入兩個整數 𝑎 和 𝑏，並判別這兩個數的大小。如果 𝑎 小於 𝑏，則印出 ‘𝑎 小於 𝑏’ （ 𝑎 和 𝑏 請用輸入的數字取代），否則印出 ‘𝑎 大於或等於𝑏’。如果使用者輸入的不是數字，則無法轉成整數，因此會有錯誤發生，此時請在 except區塊印出 ‘無法判別大小，請重新輸入’。如果 try 區塊裡沒有任何錯誤，則印出 ‘判別完成’。重複上面的步驟，直到使用者鍵入 Q 或 q 為止，然後顯示 ‘程式結束’。

```python
while True:
  s=str(input("請輸入兩個整數"))
  if(s in "Qq"):
    break
  try:
    (a,b)=s.split()
    if(a<b):
      print(f'{a}小於{b}')
    else:
      print(f'{a}大於或等於{b}')
  except Exception as e:
    print('無法判別大小，請重新輸入')
```

## 11. 試撰寫函數 to_byteArray(lst)，可以接收一個由整數組成的串列，然後將它轉換成 bytes 物件並傳回。如果無法轉換（例如整數太大、串列裡的內容是浮點數或其它原因），則顯示‘無法轉換’。

```python
def to_byteArray(lst):
  try:
    return bytes(lst)
  except:
    return '無法轉換'
print(to_byteArray([100,2,54,3]))
print(to_byteArray([100,2,54,3000]))
```
