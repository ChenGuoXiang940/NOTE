## 1.設Window類別定如下，試回答下列問題：

```python
class Window:
  def __init__(self,w=10,h=5):
    self.width=w
    self.height=h
  def area(self):
    return self.width*self.height
  def cmp(self,other):
    return max(self.area(),other.area())
#(1)試建立一個Window類別的物件𝑤0，其屬性width與height的值為預設值
w0=Window()
#(2)建立一個Window類別的物件𝑤1，並分別設定屬性width與height的值為12和8。
w1=Window(12,8)
#(3)試利用setattr() 分別將𝑤1 的屬性width與height改設為16和7。
setattr(w1,'width',16)
setattr(w1,'height',7)
print(w1.area())
#(4)試比較𝑤0和𝑤1 的面積大小（width x height）， 然後印出較大的面積。
print(w0.cmp(w1))
```

## 4.已知球體積為4/3*𝜋*r^3，表面積為4𝜋*r^2，試完成下列各題：

```python
import math
class Sphere:
  def __init__(self,r):
    self.rad=r
  def volume(self):
    return 4/3*math.pi*self.rad**3
  def surface_area(self):
    return 4*math.pi*self.rad**2
  def __repr__(self):
    return f"Sphere object, rad ={self.rad}"
  def __str__(self):
    return f"sphere object,rad={self.rad},volumn={round(self.volume(),2)},surface_area={round(self.surface_area(),2)}"
#(a)試建立一個Sphere的類別，內含一個init(r) 函數，可將rad屬性設值為r。
s0=Sphere(2)
#(b)試在Sphere類別裡定義volume()函數，可用來傳回圓球的體積，與一個surface_area() 函數，可以傳回圓球的表面積
#(c)試建立一個rad=2的Sphere類別之物件𝑠0，並求出此物件的體積和表面積
print(f'{s0.volume()} {s0.surface_area()}')

```

## 6.試設計一個Calculator類別，內含__init__(a,b) 函數，可將n1與n2屬性分別設為a與b，並設計下列各函數，然後以物件c0= Calculator(2,10) 來測試它們:

```python
import math
class Calculator:
  def __init__(self,a,b):
    self.n1=a
    self.n2=b
  def add(self):
    return self.n1+self.n2
  def gcd(self):
    return math.gcd(self.n1,self.n2)
  def lcm(self):
    return (self.n1/self.gcd())*self.n2
  def power(self):
    return (self.n1**2,self.n2**2)
obj=Calculator(2,10)
#(a)設計add()函數，可傳回n1與𝑛2之和。
print(obj.add())
#(b)設計gcd()函數，可傳回n1與n2的最大公因數
print(obj.gcd())
#(c)設計lcm()函數，可傳回n1與n2的最小公倍數。
print(obj.lcm())
#(d)設計power()函數，可傳回n1與n2次方。
print(obj.power())
```
