## 15.è©¦åˆ©ç”¨å·¢ç‹€çš„forè¿´åœˆå°å‡ºä¸‹é¢çš„åœ–æ¡ˆï¼š

```python
#(a)
for i in range(5,0,-1):
  for j in range(i):
    print('*',end='')
  print('')
#(b)
for i in range(5,-1,-1):
  for j in range(i):
    print(' ',end='')
  for j in range(5-i):
    print('*',end='')
  print('')
#(c)
for i in range(0,6,1):
  for j in range(i):
    print(' ',end='')
  for j in range(5-i):
    print('*',end='')
  print('')
#(d)
for i in range(0,6,1):
  for j in range(i):
    print('^',end='')
  for j in range(5-i):
    print('*',end='')
  print('')
#(e)
for i in range(1,7,1):
  for j in range(1,i,1):
    print(j,end='')
  print('')
#(f)
for i in range(1,7,1):
  for j in range(1,i,1):
    print(6-j,end='')
  print('')
#(g)
for i in range(0,6,1):
  for j in range(0,i,1):
    print(i,end='')
  print('')
#(h)
m=-1
for i in range(0,6,1):
  for j in range(0,i,1):
    m+=1
    if(m>9):
      print(f'{chr(m+87)}',end='')
    else:
      print(m,end='')
  print('')
```
