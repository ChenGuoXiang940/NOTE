## 5.3 While 迴圈

### 20. 試由鍵盤輸入一個整數，然後判斷輸入的數是幾個位數的整數。例如若輸入23983，則輸出 “5個位數的整數”。（提示：一個整數用整數除法除以10，其位數會少一位）。

```python
val=int(input('輸入一個整數:'))
rd=0
while int(val)!=0:
  val/=10
  rd+=1
print(rd)
```

### 21. 有時讀進來的數字會包含有千分位符號（即逗號，如13,988或1,661,231）。試由鍵盤讀入一個帶有一個或數個千分位符號的整數，然後印出這個整數乘上2之後的結果。

```python
s=input('輸入一個包含千分位的整數:')
s1=''
while len(s)!=0:
  if(s[:1].isdigit()):
    s1=s1+s[:1]
  s=s[1:]
print(int(s1)*2)
```

### 22. 整數 𝑛 的因數就是可以整除 𝑛 的數。試給予一個整數，然後列印出它所有的因數，並算出因數的個數。例如，輸入整數20，則印出 “20的因數有6個，分別為1, 2, 4, 5, 10, 20”。

```python
n=int(input('輸入一個整數n:'))
lst=[]
cnt=1
while cnt<=20:
  if(n%cnt==0):
    lst.append(cnt)
  cnt+=1
res=str(lst)
print(f'{n}的因數有{len(lst)}個，分別為{res[1:-1]}')
```

### 23. 質數是除了1和它本身之外，沒有其它因數的數，也就是說質數只會有兩個因數。請利用這個觀念來判別輸入的整數是否為質數。

```python
n=int(input('輸入一個整數'))
cnt=1
rd=0
while cnt<=n:
  if(n%cnt==0):
    rd+=1
  cnt+=1
if(rd==2):
  print(f'{n}是質數')
else:
  print(f'{n}不是質數')
```

## 5.4 break、continue與pass敘述

### 24. 試印出1到20之間，所有不能被3整除的整數。

```python
lst=[]
cnt=1
while cnt<=20:
  if(cnt%3==0):
    cnt+=1
    continue;
  lst.append(cnt)
  cnt+=1
print(lst)
```

```python
cnt=1
res=0
while res<100:
  res=cnt*(1+cnt)/2
  cnt+=1
print(cnt-1)
```

### 26. 雅筑帶一群學生去校外參訪，學生每3個一組，最後剩下2個；每5個一組，最後剩下3個；每7個一組，最後剩下2個。試問這群學生最少有多少個？試利用for迴圈求解這個問題。

```python
for cnt in range(1,100,1):
    if(cnt%3==2 and cnt%5==3 and cnt%7==2):
      print(cnt)
      break
```

### 27. 同上題，滿足上題條件的學生數有無窮多個，試找出學生數最少的3個解。

```python
cnt=1
rd=0
while True:
  if(cnt%3==2 and cnt%5==3 and cnt%7==2):
    print(cnt)
    rd+=1
  if(rd == 3):
    break
  cnt+=1
```

### 28. 設 𝑙st =[47,89,12,−4,12,2,97]，試撰寫一程式，依序讀取 𝑙st 的元素值，並將它們列印出來。如果讀到負數，則列印該負數之後，終止列印其它元素。

```python
lst=[47,89,12,-4,12,2,97]
while len(lst)!=0:
  val=lst[0]
  del lst[0]
  print(val)
  if(val<0):
    break
```

### 29. 試由鍵盤讀入一個字串（不包含單引號或雙引號），並利用迴圈逐字判別字串裡的字元是否全為數字。只要有一個字元不是數字，則輸出 “輸入的數包含不合法的字元”，否則輸出此字串。

```python
s=input('輸入一個字串:')
s1=s
while s[0].isdigit()==True:
  s=s[1:]
  if(len(s)==0):
    print(s1)
    break;
else:
  print('輸入的數包含不合法的字元')
```

### 30. 設 nums =[8,11,98,23,47]，試判別 nums內是否包含有可被3整除的數。如果有，則輸出 “包含有可被3整除的數”，否則輸出 “沒有包含可被3整除的數”。

```python
nums =[8,11,98,23,47]
while nums[0]%3!=0:
  del nums[0]
  if(len(nums)==0):
    print('沒有包含可被3整除的數')
    break
else:
  print('“包含有可被3整除的數')
```

31. 設 lower_limit =0，upper_limit =255，試由鍵盤讀入一個由整數組成的串列，然後判別此陣列內的每一個元素是否都在 lower_limit 到 upper_limit 這個範圍之內。例如，若輸入 [9,12,-1,32,-8]，則印出 “最少有一個數不在 [0, 255] 之內”，若輸入 [9, 12, 3, 32]，則印出 “輸入的數全在範圍之內”。

```python
lst=str(input('請輸入整數組成的串列')).replace(" ","").replace("[","").replace("]","").split(',')
while (int(lst[0])>= 0 and int(lst[0])<=255):
  del lst[0]
  if(len(lst)==0):
    print('輸入的數全在範圍之內')
    break
else:
  print('最少有一個數不在 [0, 255] 之內')
```

## 5.6 串列生成式 (List comprehension)

### 32. 試以串列生成式完成下列各題：

```python
#(a) 建立一個1到10的平方所組成的串列。
lst=[item**2 for item in range(1,11,1)]
print(lst)
```

```python
#(b) 建立一個1到50之間，可以同時被3和4整除的串列。
lst=[item for item in range(1,51,1) if item%3==0 and item%4==0]
print(lst)
```

```python
#(c) 取出字串 'List comprehension' 中所有的母音，並組成一個字元串列。
s='List comprehension'
target='AEIOUaeiou'
lst=[item for item in s if item in target]
print(lst)
```

```python
#(d) 依串列 [3,−1,4,7,−3,2] 的值來建立另一個新串列，若串列元素的值為正，則新串列元素的值為1，否則為 −1。本題建立出來的串列應為 [1,−1,1,1,−1,1]。
lst=[3,-1,4,7,-3,2]
res=[-1 if item <0 else 1 for item in lst]
print(res)
```

### 33. 試以串列生成式取出 ['Spring', 'Summer', 'Autumn', 'Winter'] 中所有的母音，並組成一個新的字元串列。

```python
sts=['Spring', 'Summer', 'Autumn', 'Winter']
target='AEIOUaeiou'
res=[ch for st in sts for ch in st if ch in target]
print(res)
```

### 35. 試以串列生成式建立從 num1=[1,2,3] 和 num2=[4,5] 這兩個串列中，各提取一個元素的所有組合，每一個組合以序對表示。本題輸出的結果應為 [(1, 4), (1, 5), (2, 4), (2, 5), (3, 4), (3, 5)]。

```python
num1=[1,2,3]
num2=[4,5]
res=[(item1,item2) for item1 in num1 for item2 in num2]
print(res)
```
