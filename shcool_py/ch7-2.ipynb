## 7.è¨­Employeeé¡åˆ¥å®šç¾©å¦‚ä¸‹ï¼Œå…¶ä¸­çš„é¡åˆ¥å±¬æ€§payRateæ˜¯å“¡å·¥ï¼ˆEmployeeï¼‰æ™‚è–ªçµ¦ä»˜çš„æ¯”ç‡ï¼ˆç‚ºä¸€ä¸²åˆ—ï¼‰ï¼ŒbaseSalaryæ˜¯å“¡å·¥çš„åŸºæœ¬æ™‚è–ªï¼š

### (a)è©¦å®šç¾©ä¸€å€‹å‡½æ•¸salary()ï¼Œå¯æ¥æ”¶å·¥ä½œæ™‚æ•¸hrå’Œç´¢å¼•bonusï¼Œä¸¦ä»¥payRateè£¡ç´¢å¼•ç‚ºbonusçš„å€¼åšç‚ºçµ¦ä»˜æ¯”ç‡ï¼Œç„¶å¾Œè¨ˆç®—æ‡‰çµ¦ä»˜çš„é‡‘é¡ï¼ˆå–æ•´æ•¸ï¼‰ã€‚ä¾‹å¦‚ï¼Œå¦‚æœbaseSalary=160,hr=10ï¼Œbonus=1ï¼Œå‰‡çµ¦ä»˜é‡‘é¡ç‚º160Ã—10Ã—1.2=1920ï¼Œå…¶ä¸­1.2æ˜¯ä¸²åˆ—payRateä¸­ï¼Œç´¢å¼•ç‚ºbonus=1çš„å…ƒç´ ã€‚

### (b)è©¦å®šç¾©ä¸€å€‹é¡åˆ¥å‡½æ•¸set_payRate()ï¼Œå¯ä»¥æ¥æ”¶ä¸€å€‹ä¸²åˆ—new_payRateï¼Œç„¶å¾Œå°‡é¡åˆ¥å±¬æ€§payRateä¿®æ”¹ç‚ºnew_payRateã€‚

### (c)è©¦å®šç¾©ä¸€å€‹éœæ…‹å‡½æ•¸estimate()ï¼Œå¯è¼¸å…¥åŸºæœ¬æ™‚è–ªbonusã€å·¥ä½œæ™‚æ•¸hrå’Œå’Œçµ¦ä»˜æ¯”ä¾‹rateï¼Œç„¶å¾Œä¼°ç®—å¯ç²å¾—çš„è–ªè³‡ï¼ˆå–æ•´æ•¸ï¼‰ã€‚ä¾‹å¦‚ï¼Œestimate(160,10,1.25) çš„è¨ˆç®—çµæœç‚º2000ã€‚

```python
class Employee:
  payRate=[1,1.2,1.5]    # çµ¦ä»˜çš„æ¯”ç‡
  def __init__(self,base):
    self.baseSalary=base
  def salary(self,hr,bonus):
    self.bonus=bonus
    self.hr=hr
    return self.payRate[bonus]*self.baseSalary*hr
  @classmethod
  def set_payRate(cls,new_payRate):
    cls.payRate=new_payRate
  @staticmethod
  def estimate(base,hr,rate):
    return base*hr*rate
tom=Employee(160)
print(tom.salary(10,1))       # 1920
Employee.set_payRate([1,1.3,1.5])
print(tom.salary(10,1))       # 2080
print(Employee.estimate(160,10,1.25))# 2000
```

## 8.è¨­Factorçš„é¡åˆ¥å®šç¾©å¦‚ä¸‹ï¼Œè©¦å®Œæˆå¾ŒçºŒçš„å•é¡Œï¼š

### (a)è©¦å®šç¾©ä¸€å€‹å¯¦ä¾‹å‡½æ•¸find_factors()ï¼Œç”¨ä¾†æ‰¾å‡ºfactor_listçš„å…ƒç´ ä¸­ï¼Œæœ‰å“ªå¹¾å€‹æ˜¯numçš„å› æ•¸ï¼ˆå³å¯ä»¥æ•´é™¤numçš„æ•¸ï¼‰ã€‚

### (b)è©¦å®šç¾©ä¸€å€‹é¡åˆ¥å‡½æ•¸add_factors()ï¼Œå¯æ¥æ”¶ä¸€å€‹æ•´æ•¸ä¸²åˆ—lstï¼Œç„¶å¾ŒæŠŠlstæœ‰è€Œfactor_lstæ²’æœ‰çš„å…ƒç´ åŠ åˆ°factor_lstä¸­ã€‚

### (c)è©¦å®šç¾©ä¸€å€‹é¡åˆ¥å‡½æ•¸remove_factors()ï¼Œå¯æ¥æ”¶ä¸€å€‹æ•´æ•¸ä¸²åˆ—lstï¼Œç„¶å¾Œç§»é™¤factor_lstä¸­ï¼Œæ–¼lstè£¡å‡ºç¾çš„å…ƒç´ ã€‚

### (d)è©¦å®šç¾©ä¸€å€‹é¡åˆ¥å‡½æ•¸show_factor_list()ï¼Œç”¨ä¾†é¡¯ç¤ºfactor_lstè£¡çš„å…ƒç´ ã€‚

### (e)è©¦å®šç¾©ä¸€å€‹éœæ…‹å‡½æ•¸isfactor(num,n)ï¼Œå¯ä»¥åˆ¤åˆ¥ğ‘›æ˜¯å¦ç‚ºnumçš„å› æ•¸ã€‚

```python
import math
class Factor:
  factor_list=[2,3,6,8]
  def __init__(self,num):
    print('initial factor_list:',Factor.factor_list)
    self.num=num
  def find_factors(self):
    res=[]
    for i in self.factor_list:
      if(Factor.isfactor(self.num,i)):
        res.append(i)
    print(res)
  @classmethod
  def add_factors(cls,lst):
    for i in lst:
      if((i in cls.factor_list) is False):
        cls.factor_list.append(i)
  @classmethod
  def remove_factors(cls,lst):
    for i in lst:
      if(i in cls.factor_list):
        cls.factor_list.remove(i)
  @classmethod
  def show_factor_list(cls):
    print('factor_list:',cls.factor_list)
  @staticmethod
  def isfactor(num,n):
    return True if num%n==0 else False
f0=Factor(60)       #initial factor_list: [2, 3, 6, 8]
f0.find_factors()     #[2, 3, 6]
Factor.add_factors([3,9])
Factor.show_factor_list() #factor_list: [2, 3, 6, 8, 9]
Factor.remove_factors([2,4])
Factor.show_factor_list() #factor_list: [3, 6, 8, 9]
f0.find_factors()     #[3, 6]
Factor.isfactor(45,8)   #False
```

### (a)è©¦å®šç¾©ä¸€å€‹Trucké¡åˆ¥ï¼Œå®ƒç¹¼æ‰¿è‡ªCaré¡åˆ¥ã€‚Truckçš„__init__() å‡½æ•¸å¯ä»¥æ¥æ”¶drã€owå’Œcoä¸‰å€‹åƒæ•¸   ï¼Œåˆ†åˆ¥ç”¨ä¾†è¨­å®šdoorsã€owerå’Œcolorä¸‰å€‹å¯¦ä¾‹å±¬æ€§ï¼Œå…¶ä¸­colorå±¬æ€§çš„è¨­å®šå¿…é ˆå‘¼å«çˆ¶é¡åˆ¥çš„__init__() å‡½æ•¸ã€‚

### (b)è©¦åœ¨Trucké¡åˆ¥å…§åŠ å…¥ä¸€å€‹show() å‡½æ•¸ï¼Œç”¨ä¾†åˆ—å°doorsã€ownerå’Œcolorä¸‰å€‹å¯¦ä¾‹å±¬æ€§çš„å€¼ã€‚

```python
class Car:
  def __init__(self,color):
    self.color=color
  def show(self):
    print(f'color={self.color}')
class Truck(Car):
  def __init__(self,dr,ow,co):
    super().__init__(co)
    self.doors=dr
    self.ower=ow
    self.color=co
  def show(self):
    print('doors=',self.doors,", owner=",self.ower,", color=",self.color)
Car('red').show()#color=red
Truck(2,'Tom','blue').show()#doors=2, owner=Tom, color=blue
```

## 12.è¨­Yearé¡åˆ¥çš„å®šç¾©å¦‚ä¸‹ï¼Œå…¶ä¸­çš„ç§æœ‰å±¬æ€§ __yearç‚ºè¥¿å…ƒçš„å¹´ä»½ï¼š

### (a)è©¦è¨­è¨ˆisleap() å‡½æ•¸ï¼Œå¯ä»¥åˆ¤åˆ¥ __yearæ˜¯å¦ç‚ºé–å¹´ã€‚

### (b)è¨­y0ç‚ºYearé¡åˆ¥çš„ç‰©ä»¶ï¼Œè©¦è¨­è¨ˆä¸€å€‹Getterå’ŒSetterï¼Œä½¿å¾—è¼¸å…¥y0.yearæ™‚å¯ä»¥å–å¾—__yearçš„å€¼ï¼Œ  è¼¸å…¥y0.year=yæ™‚å¯ä»¥å°‡ __yearå±¬æ€§è¨­å®šç‚ºyã€‚

```python
from sys import settrace
class Year:
  def __init__(self,y):
    self.__year=y
  def isleap(self):
    if(self.__year%400==0):
      print(True)
    elif(self.__year%100==0):
      print(False)
    elif(self.__year%4==0):
      print(True)
    else:
      print(False)
  @property
  def year(self):
    print(self.__year)
  @year.setter
  def year(self,y):
    self.__year=y
y0=Year(2022)
y0.year  #2022
y0.isleap()#False
y0.year=2020
y0.year  #2020
y0.isleap()#True
```
