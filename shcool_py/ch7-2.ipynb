## 7.設Employee類別定義如下，其中的類別屬性payRate是員工（Employee）時薪給付的比率（為一串列），baseSalary是員工的基本時薪：

### (a)試定義一個函數salary()，可接收工作時數hr和索引bonus，並以payRate裡索引為bonus的值做為給付比率，然後計算應給付的金額（取整數）。例如，如果baseSalary=160,hr=10，bonus=1，則給付金額為160×10×1.2=1920，其中1.2是串列payRate中，索引為bonus=1的元素。

### (b)試定義一個類別函數set_payRate()，可以接收一個串列new_payRate，然後將類別屬性payRate修改為new_payRate。

### (c)試定義一個靜態函數estimate()，可輸入基本時薪bonus、工作時數hr和和給付比例rate，然後估算可獲得的薪資（取整數）。例如，estimate(160,10,1.25) 的計算結果為2000。

```python
class Employee:
  payRate=[1,1.2,1.5]    # 給付的比率
  def __init__(self,base):
    self.baseSalary=base
  def salary(self,hr,bonus):
    self.bonus=bonus
    self.hr=hr
    return self.payRate[bonus]*self.baseSalary*hr
  @classmethod
  def set_payRate(cls,new_payRate):
    cls.payRate=new_payRate
  @staticmethod
  def estimate(base,hr,rate):
    return base*hr*rate
tom=Employee(160)
print(tom.salary(10,1))       # 1920
Employee.set_payRate([1,1.3,1.5])
print(tom.salary(10,1))       # 2080
print(Employee.estimate(160,10,1.25))# 2000
```

### (a)試定義一個實例函數find_factors()，用來找出factor_list的元素中，有哪幾個是num的因數（即可以整除num的數）。

### (b)試定義一個類別函數add_factors()，可接收一個整數串列lst，然後把lst有而factor_lst沒有的元素加到factor_lst中。

### (c)試定義一個類別函數remove_factors()，可接收一個整數串列lst，然後移除factor_lst中，於lst裡出現的元素。

### (d)試定義一個類別函數show_factor_list()，用來顯示factor_lst裡的元素。

### (e)試定義一個靜態函數isfactor(num,n)，可以判別𝑛是否為num的因數。

```python
import math
class Factor:
  factor_list=[2,3,6,8]
  def __init__(self,num):
    print('initial factor_list:',Factor.factor_list)
    self.num=num
  def find_factors(self):
    res=[]
    for i in self.factor_list:
      if(Factor.isfactor(self.num,i)):
        res.append(i)
    print(res)
  @classmethod
  def add_factors(cls,lst):
    for i in lst:
      if((i in cls.factor_list) is False):
        cls.factor_list.append(i)
  @classmethod
  def remove_factors(cls,lst):
    for i in lst:
      if(i in cls.factor_list):
        cls.factor_list.remove(i)
  @classmethod
  def show_factor_list(cls):
    print('factor_list:',cls.factor_list)
  @staticmethod
  def isfactor(num,n):
    return True if num%n==0 else False
f0=Factor(60)       #initial factor_list: [2, 3, 6, 8]
f0.find_factors()     #[2, 3, 6]
Factor.add_factors([3,9])
Factor.show_factor_list() #factor_list: [2, 3, 6, 8, 9]
Factor.remove_factors([2,4])
Factor.show_factor_list() #factor_list: [3, 6, 8, 9]
f0.find_factors()     #[3, 6]
Factor.isfactor(45,8)   #False
```

## 9.設Car類別定義如下：

### (a)試定義一個Truck類別，它繼承自Car類別。Truck的__init__() 函數可以接收dr、ow和co三個參數   ，分別用來設定doors、ower和color三個實例屬性，其中color屬性的設定必須呼叫父類別的__init__() 函數。

### (b)試在Truck類別內加入一個show() 函數，用來列印doors、owner和color三個實例屬性的值。

```python
class Car:
  def __init__(self,color):
    self.color=color
  def show(self):
    print(f'color={self.color}')
class Truck(Car):
  def __init__(self,dr,ow,co):
    super().__init__(co)
    self.doors=dr
    self.ower=ow
    self.color=co
  def show(self):
    print('doors=',self.doors,", owner=",self.ower,", color=",self.color)
Car('red').show()#color=red
Truck(2,'Tom','blue').show()#doors=2, owner=Tom, color=blue
```

### (b)試定義一個Student類別，它繼承自Person類別，  並設計  __init__() 函數，使得實例屬性name和gender（性別）可以被設值（name的設值請呼叫父類別的__init__()函數）。

## 11.設Triangle類別的定義如下，其中的area() 函數可用來計算底為base，高為height之三角形的面積：

## 如果先執行t0=Triangle(10,5)，再執行t0.area()，可得三角形的面積為25。但是如果接著設定t0.__base=40（base前面有兩個底線），然後執行t0.area()，得到的三角形面積卻還是25。試解釋為什麼會有這個現象，並修改程式碼加入設定私有屬性的函數，使得__base與__height也可以正確的被設定。

```python
class Triangle:
  def __init__(self,base,height):
    self.__base=base  #定義私有屬性，無法在外部做更改
    self.__height=height
  def area(self):
    return 0.5*self.__base*self.__height
t0=Triangle(10,5)
t0.base=40
print(t0.area())
```

## 12.設Year類別的定義如下，其中的私有屬性 __year為西元的年份：

### (a)試設計isleap() 函數，可以判別 __year是否為閏年。

### (b)設y0為Year類別的物件，試設計一個Getter和Setter，使得輸入y0.year時可以取得__year的值，  輸入y0.year=y時可以將 __year屬性設定為y。

```python
from sys import settrace
class Year:
  def __init__(self,y):
    self.__year=y
  def isleap(self):
    if(self.__year%400==0):
      print(True)
    elif(self.__year%100==0):
      print(False)
    elif(self.__year%4==0):
      print(True)
    else:
      print(False)
  @property
  def year(self):
    print(self.__year)
  @year.setter
  def year(self,y):
    self.__year=y
y0=Year(2022)
y0.year  #2022
y0.isleap()#False
y0.year=2020
y0.year  #2020
y0.isleap()#True
```
