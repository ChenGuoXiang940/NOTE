## 2.è©¦è¨­è¨ˆpow(ğ‘¥,ğ‘›) å‡½æ•¸ï¼Œå¯ç”¨ä¾†è¨ˆç®—ä¸¦å‚³å›ğ‘¥çš„ğ‘›æ¬¡æ–¹ã€‚

```python
import math
def pow(x,n):
  if(n==0):
    return 1
  elif(n==1):
    return x
  elif(n%2==1):
    return x*pow(x**2,n//2)
  else:
    return pow(x**2,n//2)
print(pow(2,10))
```

## 3.è©¦è¨­è¨ˆisprime(ğ‘¥) å‡½æ•¸ï¼Œç”¨ä¾†åˆ¤åˆ¥ğ‘¥æ˜¯å¦ç‚ºè³ªæ•¸ã€‚å¦‚æœæ˜¯ï¼Œå‰‡å‚³å›Trueï¼Œå¦å‰‡å‚³å›Falseã€‚

```python
def isprime(n):
  if(n==2 or n==3):
    return True
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
print(isprime(2))
print(isprime(10))
print(isprime(11))
print(isprime(13))
```

## 4.è©¦è¨­è¨ˆprimes(ğ‘¥) å‡½æ•¸ï¼Œå¯ä»¥ç”¨ä¾†æ‰¾å‡ºå°æ–¼ç­‰æ–¼ğ‘¥çš„æ‰€æœ‰è³ªæ•¸ï¼Œä¸¦å°‡æ‰¾å‡ºä¾†çš„è³ªæ•¸ä»¥ä¸²åˆ—çš„å‹å¼å‚³å›ã€‚

```python
def primes(x):
  my_list=[]
  if(x>=2):
    my_list.append(2)
  if(x>=3):
    my_list.append(3)
  for i in range(5,x+1,6):
    if(isprime(i)):
      my_list.append(i)
    if(isprime(i+2)):
      my_list.append(i+2)
  return my_list
print(primes(100))
```

## 6.è©¦è¨­è¨ˆä¸€å‡½æ•¸factor(ğ‘¥)ï¼Œå¯ä»¥å°‡æ•´æ•¸åƒæ•¸ğ‘¥é€²è¡Œè³ªå› æ•¸åˆ†è§£ï¼Œä¸¦å°‡çµæœä»¥åºå°çµ„æˆçš„ä¸²åˆ—å‚³å›ã€‚ä¾‹å¦‚ 72=23Ã—32ï¼Œå› æ­¤factor(72) å‚³å›[(2,3), (3,2)]ï¼Œåˆä¾‹å¦‚330=2 Ã—3 Ã—5 Ã—11ï¼Œå› æ­¤factor(330) å‚³å›[(2,1), (3,1), (5,1), (11,1)]ã€‚

```python
def factor(x):
  my_list=[]
  primes_list=primes(x)
  for i in primes_list:
    rd=0
    while x%i==0:
      x/=i
      rd+=1
    if(rd!=0):
      my_list.append((i,rd))
  return my_list
print(factor(72))
print(factor(330))
```

## 7.è©¦è¨­è¨ˆä¸€å‡½æ•¸bin2dec(bs)ï¼Œç”¨ä¾†å°‡äºŒé€²ä½çš„æ•¸å­—bsï¼ˆç‚ºä¸€å­—ä¸²ï¼Œè£¡é¢çš„å­—å…ƒåªèƒ½æ˜¯0æˆ–1ï¼Œbså¯ä»¥æ˜¯ä»»æ„é•·åº¦ï¼‰è½‰æ›æˆ10é€²ä½çš„æ•´æ•¸ã€‚å¦‚æœbsçš„é•·åº¦ç‚º0ï¼Œå‰‡å‚³å›Noneã€‚æç¤ºï¼š'1011' = 1 Ã—8+0 Ã—4+1 Ã—2+1 Ã—1=11ï¼Œå› æ­¤bin2dec('1011') å‚³å›11ã€‚

```python
def bin2dec(bs):
  if(len(bs)==0):
    return None
  res=0
  m=1
  while len(bs)!=0:
    if(bs[-1]=='1'):
      res+=m
    m*=2
    bs=bs[:-1]
  return res
print(bin2dec('1011'))
```

## 10.ä¸€å€‹æ•¸å¦‚æœæ˜¯æ‰€æœ‰å°æ–¼å®ƒæœ¬èº«çš„å› æ•¸ä¹‹å’Œï¼Œé€™å€‹æ•¸å°±ç¨±ç‚ºå®Œç¾æ•¸ï¼ˆPerfect  numberï¼‰ã€‚ä¾‹å¦‚6çš„å› æ•¸ä¸­ï¼Œå°æ–¼6çš„å› æ•¸æœ‰1, 2, 3ä¸”6 =1 +2 +3ï¼Œæ‰€ä»¥6æ˜¯å®Œç¾æ•¸ã€‚å¦å¤–ï¼Œ28=1 +2 +4 +7 +14ï¼Œæ‰€ä»¥28ä¹Ÿæ˜¯ä¸€å€‹å®Œç¾æ•¸ã€‚(a)è©¦æ’°å¯«ä¸€å‡½æ•¸factors(ğ‘›)ï¼Œå¯ä»¥å‚³å›ğ‘›çš„å› æ•¸ä¸­ï¼Œæ‰€æœ‰å°æ–¼ğ‘›çš„å› æ•¸ã€‚(b)è©¦åˆ©ç”¨(a) çš„çµæœæ‰¾å‡º4å€‹ä½æ•¸ä»¥ä¸‹æ‰€æœ‰çš„å®Œç¾æ•¸ã€‚

## 16.åä¸ƒä¸–ç´€çš„æ•¸å­¸å®¶æ­æ‹‰ï¼ˆEulerï¼‰æ‰¾åˆ°äº†ä¸€å€‹è¨ˆç®—åœ“å‘¨ç‡çš„å…¬å¼ï¼ˆå¥½å²å®³ï¼Œä¸çŸ¥é“ä»–æ€éº¼æ¨å°çš„ï¼‰ï¼šå…¬å¼, å…¶ä¸­æ‰€æœ‰çš„åˆ†å­éƒ½æ˜¯å¤§æ–¼2çš„è³ªæ•¸ï¼Œåˆ†æ¯å‰‡æ˜¯ä¸èƒ½è¢«4æ•´é™¤ï¼Œä¸”æœ€é è¿‘åˆ†å­çš„å¶æ•¸ã€‚è©¦æ’°å¯«ä¸€å‡½æ•¸euler_pi(ğ‘›)ï¼Œå¯ä»¥åˆ©ç”¨ä¸Šé¢çš„å…¬å¼è¨ˆç®—åœ“å‘¨ç‡åˆ°ç¬¬ğ‘›é …ã€‚ä¾‹å¦‚è‹¥ğ‘›ç‚º5ï¼Œå‰‡è¨ˆç®—åˆ°13/14é€™ä¸€é …ã€‚å¦‚æœğ‘›çœç•¥ï¼Œå‰‡é è¨­åˆ°ç¬¬10é …ã€‚

```python
import math
def isprime(n):
  if(n==2 or n==3):
    return True
  m=n%6;
  if(m!=1 and m!=5):
    return False
  sq=int(math.sqrt(n))
  for i in range(5,sq,6):
    if (n % i == 0 or n % (i + 2) == 0):
      return False
  return True
def find_down(num):
    num1=(num//2)*2
    num2=(num//2+1)*2
    if(num1%4==0):
      return num2
    else:
      return num1
def euler_pi(n=10):
  pi=2
  if(n>=1):
    pi*=(3/2)
    n-=1
  cnt=5
  while n>0:
    if(isprime(cnt)):
      pi*=(cnt/find_down(cnt))
      #print(f'{cnt} {find_down(cnt)}')
      n-=1
    if(isprime(cnt+2)):
      pi*=((cnt+2)/find_down(cnt+2))
      #print(f'{cnt+2} {find_down(cnt+2)}')
      n-=1
    cnt+=6
  return pi
print(euler_pi(10))
print(euler_pi())
```

## 17.è©¦è¨­è¨ˆä¸€å‡½æ•¸make_dict(key,value) å‡½æ•¸ï¼Œå¯ä»¥ä¾æ“šçµ¦å®šçš„keyå’Œvalueä¾†å»ºç«‹ä¸¦å‚³å›ä¸€å€‹å­—å…¸ï¼Œå…¶ä¸­keyå’Œvalueæ˜¯é•·åº¦ç›¸ç­‰çš„ä¸²åˆ—ã€‚è‹¥valueçœç•¥ï¼Œå‰‡æ‰€æœ‰keyå°æ‡‰çš„valueçš†ç‚º0ã€‚ä¾‹å¦‚make_dict([0,  1,  2],  [32,  43,  55])  å¯å‚³å›{0:  32,  1:  43,  2:  55}ï¼Œè€Œmake_dict([0, 1, 2]) å‰‡å‚³å›{0: 0, 1: 0, 2: 0}ã€‚

```python
my_dict=dict()
def make_dict(keys,values=None):
  for i in range(len(keys)):
    if values!=None and i<len(values):
      my_dict[keys[i]]=values[i]
    else:
      my_dict[keys[i]]=0
make_dict([0, 1, 2], [32, 43, 55])
print(my_dict)
my_dict.clear()
make_dict([0, 1, 2])
print(my_dict)
```
